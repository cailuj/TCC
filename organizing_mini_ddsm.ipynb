#Organização MINI-DDSM

```python
import os
from google.colab import drive
drive.mount('/content/drive')
```

## Configurar Caminhos e Pastas de Destino

```python
source_path = '/content/drive/MyDrive/MINI-DDSM'
dest_path = '/content/drive/MyDrive/organized data'

subfolders = ['benign', 'cancer', 'normal']

os.makedirs(dest_path, exist_ok=True)
print(f"Main directory ensured: {dest_path}")

for folder in subfolders:
    folder_path = os.path.join(dest_path, folder)
    os.makedirs(folder_path, exist_ok=True)
    print(f"Subfolder ensured: {folder_path}")
```

## Organização e cópias

```python
import shutil

counters = {folder: 0 for folder in subfolders}

print(f"Starting traversal of {source_path}...")

for root, dirs, files in os.walk(source_path):
    rel_path = os.path.relpath(root, source_path)
    top_folder = rel_path.split(os.sep)[0]
    category = top_folder.lower()

    if category in subfolders:
        for file in files:
            if file.lower().endswith('.jpg') and 'Mask' not in file:
                src_file_path = os.path.join(root, file)
                dest_folder_path = os.path.join(dest_path, category)

                base_name, ext = os.path.splitext(file)
                new_filename = file
                dest_file_path = os.path.join(dest_folder_path, new_filename)

                dup_count = 1
                while os.path.exists(dest_file_path):
                    new_filename = f"{base_name}_{dup_count}{ext}"
                    dest_file_path = os.path.join(dest_folder_path, new_filename)
                    dup_count += 1

                shutil.copy(src_file_path, dest_file_path)
                counters[category] += 1

print("Organizing complete.")
for category, count in counters.items():
    print(f"Total {category} images copied: {count}")
```

##Limpeza adicional

```python
dest_path = '/content/drive/MyDrive/organized data'
subfolders = ['benign', 'cancer', 'normal']

deleted_counters = {folder: 0 for folder in subfolders}

print(f"Cleaning 'mask' files from {dest_path}...")

for root, dirs, files in os.walk(dest_path):
    for file in files:
        if 'mask' in file.lower():
            file_path = os.path.join(root, file)
            try:
                os.remove(file_path)
                rel_path = os.path.relpath(root, dest_path)
                category = rel_path.split(os.sep)[0]
                if category in deleted_counters:
                    deleted_counters[category] += 1
                else:
                    print(f"Removed file from unexpected path: {rel_path}")

            except Exception as e:
                print(f"Error deleting {file_path}: {e}")

print("Cleaning complete.")
for category, count in deleted_counters.items():
    print(f"Removed {count} mask files from {category}")
```
